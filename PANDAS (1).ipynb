# **PANDAS**

Pandas in python library used for working with data sets. It has functions for analyzing, cleaning, exploring, and manipulating data.

Pandas allows us to analyze big and make conclusions based on statical theories. Pandas can clean messy data sets, and make them readable and relevant .

Relevant data is very important in data science.

In Pandas, One dimension is called Series. Two dimension is called Data Frame and Three dimension is called Panel.

# **SERIES GENERATION (1 DIMENSION)**

```python
#SERIES GENERATION

import pandas as pd

data = [3,4,5,8]

s = pd.Series(data)

print(s)
```

```python
#arrays in series

import pandas as pd
import numpy as np

data = np.array([4,5,6,7])

print(type(data),'\n')

s = pd.Series(data)

print(s)
```

```python
import pandas as pd
import numpy as np

data = np.array([[4,5,6,7],[1,2,3,4]])

s = pd.Series(data)

print(s)
```

```python
#LIST

import pandas as pd

import numpy as np

data1 = [1,3,5,7,9]

print(type(data1),'\n')

s = pd.Series(data1)

print(s)
```

```python
#TUPLE

import pandas as pd

data2 = (10, 20, 30, 40, 50)

print(type(data2),'\n')

series = pd.Series(data2)

print(series)
```

```python

import pandas as pd

data = (4,5,6,7)

s = pd.Series(data)

print(s,'\n')

print('Shape : ',s.shape)

print('Dimension : ',s.ndim)

print('D',s.dtype)
```

# **Custom Indexing**

```python
#INDEXING

import pandas as pd

import numpy as np

data = ([4,5,6,7])

s = pd.Series(data,index = [100,101,102,103])

print(s)
```

```python
#DICTIONARY

import pandas as pd

data = {1:'shaluu',2:'nasrn',3:'fathm'}

s = pd.Series(data)

print(s)
```

```python
#INDEXING


import pandas as pd

import numpy as np

data = {1:'shahla',2:'fathm',3:'nsrn'}

s = pd.Series(data,index = [5,1,8,2,3,4])

print(s)
```

```python
#INDEXING A PARTICULAR VALUE

import pandas as pd

import numpy as np

data = {1:'shahla',2:'fathm',3:'nsrn'}

s = pd.Series(data,index = [5,1,8,2,3,4])

print(s[3])
```

```python
#INDEXING MULTIPLE VALUES

import pandas as pd

import numpy as np

data = {1:'shahla',2:'fathm',3:'nsrn'}

s = pd.Series(data,index = [5,1,8,2,3,4])

print(s[[1,2,3]])
```

```python
#SLICING

import pandas as pd

import numpy as np

data = {1:'shahla',2:'fathm',3:'nsrn'}

s = pd.Series(data,index = [5,1,8,2,3,4])

print(s[0:3])
```

```python
import pandas as pd
import numpy as np

mark = {'Name': ['Shahla','Fathima','Nasrin'],'Maths': [40,60,50],'Physics': [40,60,50],'Chemistry' : [40,60,50]}

x = pd.DataFrame(mark)

x['Biology'] = [44,50,60]

x['Total Mark'] = x['Maths'] + x['Physics'] + x['Chemistry'] + x['Biology']

list1 = []


for i in x['Total Mark']:


  if i >= 200:

    list1.append('Excellent')

  elif i > 100:
    list1.append('Average')

  else:
    list1.append('Below Average')


x['Status'] = list1

print(x)

print(x[['Name','Status','Chemistry']])
```

```python
import pandas as pd
import numpy as np

mark = {'Name': ['Shahla','Fathima','Nasrin'],'Maths': [40,60,50],'Physics': [40,60,50],'Chemistry' : [40,60,50]}

x = pd.DataFrame(mark)

x['Biology'] = [44,50,60]

x['Total Mark'] = x['Maths'] + x['Physics'] + x['Chemistry'] + x['Biology']

list1 = []


for i in x['Total Mark']:


  if i >= 200:

    list1.append('Excellent')

  elif i > 100:
    list1.append('Average')

  else:
    list1.append('Below Average')


x['Status'] = list1

del x['Chemistry']

x = x.drop('Physics', axis = 1) #want to specify axis

x.pop('Maths')

print(x)
```

```python
import pandas as pd
import numpy as np

mark = {'Name': ['Shahla','Fathima','Nasrin'],'Maths': [40,60,50],'Physics': [40,60,50],'Chemistry' : [40,60,50]}

x = pd.DataFrame(mark)

x['Biology'] = [44,50,60]

x['Total Mark'] = x['Maths'] + x['Physics'] + x['Chemistry'] + x['Biology']

list1 = []


for i in x['Total Mark']:


  if i >= 200:

    list1.append('Excellent')

  elif i > 100:
    list1.append('Average')

  else:
    list1.append('Below Average')


x['Status'] = list1

print(x)

print(x.ndim)
print('\n')
print(x.shape)
print('\n')
print(x.size)
print('\n')
print(x.columns)
print('\n')
print(x['Status'].unique())
print('\n')
print(x['Biology'].unique())
print('\n')
print(x['Status'].value_counts())
```
